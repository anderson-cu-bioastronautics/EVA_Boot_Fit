```python
import numpy as np
from matplotlib import pyplot as plt
import scipy.stats
```

```python
rawData = np.genfromtxt('rawData.csv',delimiter=',')
filteredData = np.genfromtxt('filteredData.csv',delimiter=',')
zvuData = np.genfromtxt('zeroVelUpdateData.csv',delimiter=',')
```

```python
rawData
```

```python
cols = ['Human', 'Spacesuit']
rows = ['Raw Data', 'Filtered Data', 'ZVU Data']

fig, ax = plt.subplots(3,2,figsize=(12,8))
ax[0,0].hist(rawData[:,9]/rawData[:,5])
ax[0,1].hist(rawData[:,10]/rawData[:,6])
ax[1,0].hist(filteredData[:,9]/filteredData[:,5])
ax[1,1].hist(filteredData[:,10]/filteredData[:,6])
ax[2,0].hist(zvuData[:,9]/zvuData[:,5])
ax[2,1].hist(zvuData[:,10]/zvuData[:,6])

for axes, col in zip(ax[0], cols):
    axes.set_title(col)
    
for axes, row in zip(ax[:,0], rows):
    axes.set_ylabel(row, size='large')

ax[0,0].set_ylim([0,500])
ax[0,1].set_ylim([0,500])

ax[1,0].set_ylim([0,500])
ax[1,1].set_ylim([0,500])

ax[2,0].set_ylim([0,500])
ax[2,1].set_ylim([0,500])


plt.suptitle('Stance Phase Drift')
plt.show()
```

```python
#Change to Drift Rates

stanceH = np.asarray([rawData[:,9]/rawData[:,5],filteredData[:,9]/filteredData[:,5],zvuData[:,9]/zvuData[:,5]])
stanceSS = np.asarray([rawData[:,10]/rawData[:,6],filteredData[:,10]/filteredData[:,6],zvuData[:,10]/zvuData[:,6]])

swingH = np.asarray([rawData[:,11]/rawData[:,7],filteredData[:,11]/filteredData[:,7],zvuData[:,11]/zvuData[:,7]])
swingSS = np.asarray([rawData[:,12]/rawData[:,8],filteredData[:,12]/filteredData[:,8],zvuData[:,12]/zvuData[:,8]])
```

```python
#Normality test
print(scipy.stats.shapiro(stanceH[0,:]))
print(scipy.stats.shapiro(stanceH[1,:]))
print(scipy.stats.shapiro(stanceH[2,:]))

print(scipy.stats.shapiro(stanceSS[0,:]))
print(scipy.stats.shapiro(stanceSS[1,:]))
print(scipy.stats.shapiro(stanceSS[2,:]))

print(scipy.stats.shapiro(swingH[0,:]))
print(scipy.stats.shapiro(swingH[1,:]))
print(scipy.stats.shapiro(swingH[2,:]))

print(scipy.stats.shapiro(swingSS[0,:]))
print(scipy.stats.shapiro(swingSS[1,:]))
print(scipy.stats.shapiro(swingSS[2,:]))
```
